```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
cancer_data = datasets.load_breast_cancer()
```

```python
cancer_data
```

```python
new_dataset=pd.DataFrame(cancer_data.data,columns=cancer_data.feature_names)
```

```python
new_dataset
```

```python
new_dataset.head()
new_dataset.tail()
new_dataset.describe()
new_dataset.info()
```

```python
new_dataset.isna().sum()
```

```python
new_dataset.columns
```

```python
new_dataset['Label'] = cancer_data.target
```

```python
new_dataset
```

```python
plt.figure(figsize=(4,3))
plt.title("Distribution of target variable")
sns.countplot(x=new_dataset['Label'])


plt.show()


```

```python
new_dataset.corr()
```

```python
plt.figure(figsize=(12,12))
plt.title("Correlation Matrix")
sns.heatmap(new_dataset.corr(), annot=True, fmt='.1f', annot_kws={'size':8})
```

```python
#splitting the teams 
x=new_dataset.drop(['Label'], axis=1)
y=new_dataset['Label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=123)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
logistic = LogisticRegression()
```

```python
logistic.fit(x_train, y_train)
```

```python
logistic.coef_
```

```python
logistic.intercept_
```

```python
y_pred = logistic.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,precision_score,recall_score,f1_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
precision_score(y_test, y_pred)
```

```python
f1_score(y_test, y_pred)
```

```python
recall_score(y_test, y_pred)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred)
```
